```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.metrics import accuracy_score

# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()
```

```python
def generate_data(n_samples, noise, t1=6, t2=8, t3=9, low=0, high=10):
    X1 = np.random.uniform(low, high, size=(n_samples,1))
    X2 = np.random.choice([0, 1], size=(n_samples,1))
    X = np.concatenate((X1, X2), axis=1)

    # labels generated according to thresholds
    y = np.zeros(n_samples)
    y[np.where((X[:,0] > t1) & (X[:,0] < t2))] = 1
    y[np.where(X[:, 0] > t3)] = 1

    # Groups
    g1_disjoint = np.where((X[:,0] > 0) & (X[:,0] < t3) & (X[:,1] == 0.))
    g1g2 = np.where((X[:,0] > t2) & (X[:,0] < t3))
    g2_disjoint = np.where((X[:,0] > 0) & (X[:,0] < t3) & (X[:,1] == 1.))

    g1 = np.union1d(g1_disjoint, g1g2)
    g2 = np.union1d(g2_disjoint, g1g2)
    g3 = np.where((X[:,0] > t2))[0]
    g1g2g3 = np.intersect1d(g1g2, g3)

    # Disagreement region
    # Vary between random noise and completely deterministic
    y[g1g2g3] = np.random.choice([0, 1], size=len(g1g2g3), p=[1-noise, noise])
    return X, y, [g1,g2,g3,g1g2,g1g2g3]
```

```python
import numpy as np

def clt_bound(n:int, e:float):
    """Return the lower and upper bound on error rate when test set size is n and empirical error rate is e"""
    assert e >= 0. and e <= 1 and n >= 0, f'Invalid input: n={n}, e={e}'
    a = 4.+n
    b = 2.+n*e
    c = n*e**2
    d = 2.*np.sqrt(1.+n*e*(1.-e))
    return ((b-d)/a, (b+d)/a)
```

In the following sanity checks, the reported errors are all in-sample.

## Experiment 1: Vary noise level => Training Error Gap

```python
# Experiment setup parameters
noise_lvls = np.linspace(0, 0.5, num=50)
n_samples = 1000
bootstraps = 10
np.random.seed(40)

# parameters for trees
max_depth = 4
tree_params = {
    'max_depth': max_depth
}
rf_params = {
    'n_estimators': 50,
    'max_depth': max_depth
}
```

```python
avg_agree = []
avg_agreeg1g2 = []

avgacc_rfall = []
avgacc_rfg1 = []
avgacc_rfg2 = []
avgacc_rfg3 = []

avgacc_h1g1 = []
avgacc_h2g2 = []
avgacc_h3g3 = []

for noise in noise_lvls:
    # generate a new sample of data to get a bootstrap avg on the agreement
    agree_g1g2 = []
    agree_g1g2g3 = []

    acc_rfall = []
    acc_rfg1 = []
    acc_rfg2 = []
    acc_rfg3 = []
    
    acc_h1g1 = []
    acc_h2g2 = []
    acc_h3g3 = []
    for i in range(bootstraps):
        X, y, [g1,g2,g3,g1g2,g1g2g3] = generate_data(n_samples=n_samples,
                                                    noise=noise)
        
        # Train classifiers
        h1 = DecisionTreeClassifier(**tree_params)
        h1.fit(X[g1], y[g1])
        h2 = DecisionTreeClassifier(**tree_params)
        h2.fit(X[g2], y[g2])
        h3 = DecisionTreeClassifier(**tree_params)
        h3.fit(X[g3], y[g3])
        rf = RandomForestClassifier(**rf_params)
        rf.fit(X, y)

        # Get agreement
        agree_g1g2.append(accuracy_score(h1.predict(X[g1g2]), 
                                               h2.predict(X[g1g2])))
        agree_g1g2g3.append(accuracy_score(h1.predict(X[g1g2g3]),
                                                 h3.predict(X[g1g2g3])))
        
        # Get training accuracies
        acc_rfall.append(accuracy_score(rf.predict(X), y))
        acc_rfg1.append(accuracy_score(rf.predict(X[g1]), y[g1]))
        acc_rfg2.append(accuracy_score(rf.predict(X[g2]), y[g2]))
        acc_rfg3.append(accuracy_score(rf.predict(X[g3]), y[g3]))

        acc_h1g1.append(accuracy_score(h1.predict(X[g1]), y[g1]))
        acc_h2g2.append(accuracy_score(h2.predict(X[g2]), y[g2]))
        acc_h3g3.append(accuracy_score(h3.predict(X[g3]), y[g3]))

    # Get average agreement
    avg_agreeg1g2.append(sum(agree_g1g2)/ len(agree_g1g2))
    avg_agree.append(sum(agree_g1g2g3)/len(agree_g1g2g3))
    
    # Get average accuracies
    avgacc_rfall.append(sum(acc_rfall)/len(acc_rfall))
    avgacc_rfg1.append(sum(acc_rfg1)/len(acc_rfg1))
    avgacc_rfg2.append(sum(acc_rfg2)/len(acc_rfg2))
    avgacc_rfg3.append(sum(acc_rfg3)/len(acc_rfg3))

    avgacc_h1g1.append(sum(acc_h1g1)/len(acc_h1g1))
    avgacc_h2g2.append(sum(acc_h2g2)/len(acc_h2g2))
    avgacc_h3g3.append(sum(acc_h3g3)/len(acc_h3g3))
```

```python
# Plot 1: Agreement
fig, ax = plt.subplots()
ax.plot(noise_lvls, avg_agreeg1g2, label='$g_1 \cap g_2$')
ax.plot(noise_lvls, avg_agree, label='$g_1 \cap g_2 \cap g_3$')
ax.set_title('Multi-group agreement by noise level')
ax.legend()
plt.show()
```

```python
# Plot 2: Error of h3 vs rf on g3
fig, axs = plt.subplots(1, 3, figsize=(10, 3), layout='constrained')

# g1
axs[0].plot(noise_lvls, avgacc_h1g1, 
        label='$\hat{\mathrm{err}}(\hat{h}_1 \mid g_1)$')
axs[0].plot(noise_lvls, avgacc_rfg1, 
        label='$\hat{\mathrm{err}}(RF \mid g_1)$')
axs[0].set_title('Train acc. gap on $g_1$')
axs[0].legend()

# g2
axs[1].plot(noise_lvls, avgacc_h2g2, 
        label='$\hat{\mathrm{err}}(\hat{h}_2 \mid g_2)$')
axs[1].plot(noise_lvls, avgacc_rfg2, 
        label='$\hat{\mathrm{err}}(RF \mid g_2)$')
axs[1].set_title('Train acc. on $g_2$')
axs[1].legend()

# g3
axs[2].plot(noise_lvls, avgacc_h3g3, 
        label='$\hat{\mathrm{err}}(\hat{h}_3 \mid g_3)$')
axs[2].plot(noise_lvls, avgacc_rfg3, 
        label='$\hat{\mathrm{err}}(RF \mid g_3)$')
axs[2].set_title('Train acc. on $g_3$')
axs[2].legend()
plt.show()
```

## Experiment 2: Vary noise level => Test Error Gap

```python
# Experiment setup parameters
noise_lvls = np.linspace(0, 0.5, num=50)
n_samples = 1000
bootstraps = 50
np.random.seed(40)

# parameters for trees
max_depth = 4
tree_params = {
    'max_depth': max_depth
}
rf_params = {
    'n_estimators': 50,
    'max_depth': max_depth
}
```

```python
X, y, [g1,g2,g3,g1g2,g1g2g3] = generate_data(n_samples=n_samples,
                                                    noise=noise)
multiplier = len(X)/len(X[g3])
X, y, [g1,g2,g3,g1g2,g1g2g3] = generate_data(n_samples=int(5*n_samples),
                                                    noise=noise)
```

```python
avg_agree = []
avg_agreeg1g2 = []

avgacc_rfall = []
avgacc_rfg1 = []
avgacc_rfg2 = []
avgacc_rfg3 = []

avgacc_h1g1 = []
avgacc_h2g2 = []
avgacc_h3g3 = []

conf_rfall = []
conf_rfg1 = []
conf_rfg2 = []
conf_rfg3 = []
conf_h1 = []
conf_h2 = []
conf_h3 = []

for noise in noise_lvls:
    # generate a new sample of data to get a bootstrap avg on the agreement
    agree_g1g2 = []
    agree_g1g2g3 = []

    acc_rfall = []
    acc_rfg1 = []
    acc_rfg2 = []
    acc_rfg3 = []
    
    acc_h1g1 = []
    acc_h2g2 = []
    acc_h3g3 = []
    for i in range(bootstraps):
        X, y, [g1,g2,g3,g1g2,g1g2g3] = generate_data(n_samples=n_samples,
                                                    noise=noise)
        X_test, y_test, [g1_test,g2_test, g3_test, g1g2_test, g1g2g3_test] = generate_data(n_samples=n_samples, noise=noise)
        
        # Train classifiers
        rf = RandomForestClassifier(**rf_params)
        rf.fit(X, y)

        X, y, [g1,g2,g3,g1g2,g1g2g3] = generate_data(n_samples=10*n_samples,
                                                    noise=noise)
        h1 = DecisionTreeClassifier(**tree_params)
        h1.fit(X[g1], y[g1])
        h2 = DecisionTreeClassifier(**tree_params)
        h2.fit(X[g2], y[g2])
        h3 = DecisionTreeClassifier(**tree_params)
        h3.fit(X[g3], y[g3])

        # Get agreement
        agree_g1g2.append(accuracy_score(h1.predict(X[g1g2]), 
                                               h2.predict(X[g1g2])))
        agree_g1g2g3.append(accuracy_score(h1.predict(X[g1g2g3]),
                                                 h3.predict(X[g1g2g3])))
        
        # Get training accuracies
        acc_rfall.append(accuracy_score(rf.predict(X_test), y_test))
        acc_rfg1.append(accuracy_score(rf.predict(X_test[g1_test]), 
                                       y_test[g1_test]))
        acc_rfg2.append(accuracy_score(rf.predict(X_test[g2_test]), 
                                       y_test[g2_test]))
        acc_rfg3.append(accuracy_score(rf.predict(X_test[g3_test]), 
                                       y_test[g3_test]))

        acc_h1g1.append(accuracy_score(h1.predict(X_test[g1_test]), 
                                       y_test[g1_test]))
        acc_h2g2.append(accuracy_score(h2.predict(X_test[g2_test]), 
                                       y_test[g2_test]))
        acc_h3g3.append(accuracy_score(h3.predict(X_test[g3_test]), 
                                       y_test[g3_test]))

    # Get average agreement
    avg_agreeg1g2.append(sum(agree_g1g2)/ len(agree_g1g2))
    avg_agree.append(sum(agree_g1g2g3)/len(agree_g1g2g3))

    # Get bootstrap estimates of std. error
    conf_rfall.append(np.sqrt(np.var(acc_rfall)))
    conf_rfg1.append(np.sqrt(np.var(acc_rfg1)))
    conf_rfg2.append(np.sqrt(np.var(acc_rfg2)))
    conf_rfg3.append(np.sqrt(np.var(acc_rfg3)))

    conf_h1.append(np.sqrt(np.var(acc_h1g1)))
    conf_h2.append(np.sqrt(np.var(acc_h2g2)))
    conf_h3.append(np.sqrt(np.var(acc_h3g3)))
    
    # Get average accuracies
    avgacc_rfall.append(sum(acc_rfall)/len(acc_rfall))
    avgacc_rfg1.append(sum(acc_rfg1)/len(acc_rfg1))
    avgacc_rfg2.append(sum(acc_rfg2)/len(acc_rfg2))
    avgacc_rfg3.append(sum(acc_rfg3)/len(acc_rfg3))

    avgacc_h1g1.append(sum(acc_h1g1)/len(acc_h1g1))
    avgacc_h2g2.append(sum(acc_h2g2)/len(acc_h2g2))
    avgacc_h3g3.append(sum(acc_h3g3)/len(acc_h3g3))
```

```python
# Plot 1: Agreement
fig, ax = plt.subplots()
ax.plot(noise_lvls, avg_agreeg1g2, label='$g_1 \cap g_2$')
ax.plot(noise_lvls, avg_agree, label='$g_1 \cap g_2 \cap g_3$')
ax.set_title('Multi-group agreement by noise level')
ax.legend()
plt.show()
```

```python
# Plot 2: Error of h3 vs rf on g3
fig, axs = plt.subplots(1, 3, figsize=(10, 3), layout='constrained')

# g1
axs[0].plot(noise_lvls, avgacc_h1g1, 
        label='$\mathrm{err}(\hat{h}_1 \mid g_1)$')
axs[0].fill_between(noise_lvls, 
                    np.array(avgacc_h1g1) - 2 * np.array(conf_h1), 
                    np.array(avgacc_h1g1) + 2 * np.array(conf_h1),
                    alpha=0.2, antialiased=True)
axs[0].plot(noise_lvls, avgacc_rfg1, 
        label='$\mathrm{err}(RF \mid g_1)$')
axs[0].fill_between(noise_lvls, 
                    np.array(avgacc_rfg1) - 2 * np.array(conf_rfg1), 
                    np.array(avgacc_rfg1) + 2 * np.array(conf_rfg1),
                    alpha=0.2, antialiased=True)
axs[0].set_title('Test acc. gap on $g_1$')
axs[0].legend()

# g2
axs[1].plot(noise_lvls, avgacc_h2g2, 
        label='$\mathrm{err}(\hat{h}_2 \mid g_2)$')
axs[1].fill_between(noise_lvls, 
                    np.array(avgacc_h2g2) - 2 * np.array(conf_h2), 
                    np.array(avgacc_h2g2) + 2 * np.array(conf_h2),
                    alpha=0.2, antialiased=True)
axs[1].plot(noise_lvls, avgacc_rfg2, 
        label='$\mathrm{err}(RF \mid g_2)$')
axs[1].fill_between(noise_lvls, 
                    np.array(avgacc_rfg2) - 2 * np.array(conf_rfg2), 
                    np.array(avgacc_rfg2) + 2 * np.array(conf_rfg2),
                    alpha=0.2, antialiased=True)
axs[1].set_title('Test acc. gap on $g_2$')
axs[1].legend()

# g3
axs[2].plot(noise_lvls, avgacc_h3g3, 
        label='$\mathrm{err}(\hat{h}_3 \mid g_3)$')
axs[2].fill_between(noise_lvls, 
                    np.array(avgacc_h3g3) - 2 * np.array(conf_h3), 
                    np.array(avgacc_h3g3) + 2 * np.array(conf_h3),
                    alpha=0.2, antialiased=True)
axs[2].plot(noise_lvls, avgacc_rfg3, 
        label='$\mathrm{err}(RF \mid g_3)$')
axs[2].fill_between(noise_lvls, 
                    np.array(avgacc_rfg3) - 2 * np.array(conf_rfg3), 
                    np.array(avgacc_rfg3) + 2 * np.array(conf_rfg3),
                    alpha=0.2, antialiased=True)
axs[2].set_title('Test acc. gap on $g_3$')
axs[2].legend()
plt.show()
```
